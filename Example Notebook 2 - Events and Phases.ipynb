```python
from pystp import STPClient
%matplotlib inline
```

```python
stp = STPClient('lefroy.gps.caltech.edu', 9999)
```

```python
stp.connect()
```

# Finding Phases and Events by ID

```python
stp.get_phases(evids=[38904823])
```

Phases can be saved on disk in an output file.

```python
stp.get_phases(evids=[38904823], output_file='phases.txt')
```

Commands preceded with "!" are OS shell commands. To see the contents of the output file, you can use the `cat` command.

```python
!cat phases.txt
```

`get_events` can find information about events by ID. Like `get_phases`, an output file can be specified.

```python
stp.get_events(evids=[38904823])
```

```python
stp.get_events(evids=[38904823], output_file='events.txt')
```

```python
!cat events.txt
```

# Searching for Events

If you don't have event ID, you can search for events and phase picks by date and time, agmnitude, location, earthquake types, and geographical location types. Search parameters are passed to `get_events` and `get_phases` as lists that specify the start and end of a range. Datetimes must be objects of Python's `datetime.datetime` class.

```python
# Import datetime
from datetime import datetime
```

```python
# Create a start time and end time for our query.
t0 = datetime(year=2019, month=7, day=4, hour=0, minute=0, second=0)
t1 = datetime(year=2019, month=7, day=5, hour=0, minute=0, second=0)
```

```python
# Get local earthquakes that occurred between 2019/07/04,00:00:00 and 2019/07/05, 00:00:00,
# with magnitudes between 4 and 8. This coincides with the Ridgecrest sequence.
stp.get_events(times=[t0, t1], mags=[4, 8], types=['eq'], gtypes=['l'])
```

```python
# Get phase picks for the same earthquakes.
stp.get_phases(times=[t0, t1], mags=[4, 8], types=['eq'], gtypes=['l'])
```

```python
# Save the phase picks in an output file.
stp.get_phases(times=[t0, t1], mags=[4, 8], types=['eq'], gtypes=['l'], output_file='phases.txt')
```

```python
!cat phases.txt
```
